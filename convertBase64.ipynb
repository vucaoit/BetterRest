```python
import base64
import json
from google.colab import files

def decode_base64_keys_in_list(json_string, keys_to_decode):
    json_data = json.loads(json_string)

    # Iterate through each object in the list
    for obj in json_data:
        for key in keys_to_decode:
            if key in obj and isinstance(obj[key], str):
                try:
                    # Decode the Base64 value
                    decoded_value = base64.b64decode(obj[key]).decode('utf-8')
                    obj[key] = decoded_value
                except Exception as e:
                    print(f"Error decoding key '{key}' in object: {e}")
    return json_data

def remove_key_from_json(data, key_to_remove):
    """
    Recursively removes a specified key from a JSON object (dict or list).

    Args:
        data (dict or list): The JSON data as a Python dictionary or list.
        key_to_remove (str): The key to remove from the JSON.

    Returns:
        dict or list: The updated JSON data with the specified key removed.
    """
    if isinstance(data, dict):
        # Create a new dictionary without the key to remove
        return {key: remove_key_from_json(value, key_to_remove)
                for key, value in data.items() if key != key_to_remove}

    elif isinstance(data, list):
        # Apply the removal process for each element if it's a list
        return [remove_key_from_json(item, key_to_remove) for item in data]

    # Return the item if it's neither a dictionary nor a list
    return data


# Example usage
json_string = input("Paste your json: ")

# Keys to decode
keys_to_decode = ["response-payload-base64", "request-payload-base64"]

# Decode the specified Base64 keys in each object in the list
decoded_json_list = decode_base64_keys_in_list(json_string, keys_to_decode)
# Key to remove
key = "response-headers"

# Remove the key from the JSON data
updated_json = remove_key_from_json(decoded_json_list, key)
remove_text = json.dumps(updated_json) #.replace('"{',"{").replace('}"',"}")
remove_text = remove_text.replace('''\\n''','').replace('''\\"''','\"').replace('''"{''','{').replace('''}"''','}')
result = json.loads(remove_text)
# Save to a JSON file
with open('data.json', 'w') as json_file:
    json.dump(result, json_file, indent=4)

# Download the JSON file
files.download('data.json')
```
